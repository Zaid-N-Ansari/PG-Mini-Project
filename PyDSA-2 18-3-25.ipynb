```python
class Node:
    def __init__(self, val):
        self.val = val
        self.nxt:Node = None

class SinglyLinkedList:
    def __init__(self):
        self.head:Node = None
    
    def insert(self, val):
        a = self.head
        nn:Node = Node(val)
        if self.head is None:
            self.head = nn
        else:
            while a.nxt is not None:
                a = a.nxt
            a.nxt = nn
    
    def insert_at_beginning(self, val):
        nb:Node = Node(val)
        if self.head is None:
            self.head = nb
        else:
            a = self.head
            self.head = nb
            nb.nxt = a
    
    def insert_at_end(self, val):
        self.insert(val)

    def insert_at(self, pos, val):
        p = 0
        a = self.head
        if a is None:
            self.insert_at_beginning(val)
        else:
            while (pos-1) != p and a is not None:
                p += 1
                a = a.nxt
            nn:Node = Node(val)
            nn.nxt = a.nxt
            a.nxt = nn
    
    def del_at_pos(self, pos):
        p = 0
        a:Node = self.head
        if a is None:
            print('Singly Linked List is Empty')
        else:
            while (pos-1) != p and a is not None:
                p += 1
                a = a.nxt
            a.nxt = a.nxt.nxt

    def display(self):
        a = self.head
        if a is None:
            print('Singly Linked List is Empty')
        while a is not None:
            print(f'{a.val}{' -> ' if a.nxt is not None else ''}', end='')
            a = a.nxt

sll = SinglyLinkedList()
sll.insert(1)
sll.insert(11)
sll.insert(12)
sll.insert(13)
sll.insert(14)
sll.insert(15)
sll.insert_at(2, 100)
sll.display()
sll.del_at_pos(2)
print()
sll.display()
```

```python
class Node:
    def __init__(self, val):
        self.prv:Node = None
        self.val = val
        self.nxt:Node = None

class DoublyLinkedList:
    def __init__(self):
        self.head:Node = None

    def insert(self, val):
        nn:Node = Node(val)
        if self.head is None:
            self.head = nn
            nn.prv = self.head
        else:
            a = self.head
            while a.nxt is not None:
                a = a.nxt
            a.nxt = nn
            nn.prv = a

    def insert_at_end(self, val):
        self.insert(val)

    def insert_at_beginning(self, val):
        nb:Node = Node(val)
        if self.head is None:
            self.head = nb
            nb.prv = self.head
        else:
            a = self.head
            self.head = nb
            nb.prv = self.head
            nb.nxt = a
            a.nxt.prv = nb

    def display(self):
        a = self.head
        if a is None:
            print('Doubly Linked List is Empty')
        while a is not None:
            print(f'{a.val}{' <-> ' if a.nxt is not None else ''}', end='')
            a = a.nxt

dll = DoublyLinkedList()
dll.insert(1)
dll.insert(11)
dll.insert(12)
dll.insert(13)
dll.insert(14)
dll.insert(15)
dll.insert_at_beginning(100)
dll.insert_at_end(56)
dll.display()
```
