```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv', header=None)
# data = data.sample(100)
```

```python
data.drop([0], axis=1, inplace=True)
data
```

```python
transactions = data.apply(lambda row: row.dropna().astype(str).tolist(), axis=1)
transaction_ids = range(1, len(transactions) + 1)

df = pd.DataFrame({
    'TransactionID': transaction_ids,
    'Items': transactions
})
df
```

```python
df = df[df['Items'].map(len) > 1]

df.set_index('TransactionID', inplace=True)

df
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

item_list = df['Items'].tolist()
item_list
```

```python
te = TransactionEncoder()
te
```

```python
te_ary = te.fit(item_list).transform(item_list)
te_ary
```

```python
print(*te.columns_mapping_.items(), sep=', ')
```

```python
apriori_df = pd.DataFrame(te_ary, columns=te.columns_)
apriori_df
```

```python
frequent_itemsets = apriori(apriori_df, min_support=0.02, use_colnames=True, max_len=3)
frequent_itemsets.sort_values(by='support', ascending=False, inplace=True)
frequent_itemsets
```

```python
assoc_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4).sort_values(by='confidence', ascending=False)
assoc_rules
```

```python
assoc_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
assoc_rules
```
