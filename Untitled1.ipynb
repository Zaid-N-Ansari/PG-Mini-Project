```python
import string
from collections import Counter
from nltk.corpus import stopwords

data = ''
for txt in open('./alchemist.txt', 'r').readlines():
    data += txt
text = data.split('\n')

cleaned_text = []
for line in text:
    cleaned_line = line.translate(str.maketrans("", "", string.punctuation)).lower()
    cleaned_text.append(cleaned_line)

tokens = []
for line in cleaned_text:
    tokens.extend(line.split())

stop_words = set(stopwords.words('english'))
counter = Counter(tokens)

total_tokens = sum(counter.values())
unique_words = set(tokens)
total_types = len(unique_words)

stop_word_types = len([word for word in unique_words if word in stop_words])
non_stop_word_types = len([word for word in unique_words if word not in stop_words])

print(f'Total Types are: {total_types}')
print(f'Total Tokens are: {total_tokens}')
print(f'TTR (Type-Token Ratio): {round(total_types / total_tokens, 5)}')

print(f'Total Stop Word Types are: {stop_word_types}')
print(f'Total Non-Stop Word Types are: {non_stop_word_types}')
```

```python
import re
from collections import Counter, defaultdict
import math # For log probabilities if needed, though not used in basic P(w|...)

# --- Step 1: Reading Text (or defining sample text) ---
# Option 1: Define sample text directly
text_data = """
N-gram modeling is a fundamental technique.
Text analysis benefits from n-gram models.
Understanding n-grams helps in understanding language structure.
Probabilistic models use n-gram frequencies.
"""

print("--- Original Text ---")
print(text_data)

# --- Step 2: Text Preprocessing ---

def preprocess_text(text):
    """
    Tokenizes, lowercases, and removes punctuation from text.
    """
    # Lowercase
    text = text.lower()
    # Remove punctuation (keeps alphanumeric and spaces)
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize (simple split by whitespace)
    tokens = text.split()
    return tokens

print("\n--- Preprocessing ---")
processed_tokens = preprocess_text(text_data)
print("Tokens:", processed_tokens)

# --- Step 3: Generate N-grams ---

def generate_ngrams(tokens, n):
    """
    Generates n-grams from a list of tokens.
    """
    ngrams = []
    for i in range(len(tokens) - n + 1):
        # Extract sequence of n tokens
        ngram = tuple(tokens[i : i + n])
        ngrams.append(ngram)
    return ngrams

print("\n--- Generating N-grams ---")
n_value = 2 # Example: Generate Bigrams
ngrams_list = generate_ngrams(processed_tokens, n_value)
print(f"{n_value}-grams (Bigrams):", ngrams_list)

n_value_tri = 3 # Example: Generate Trigrams
trigrams_list = generate_ngrams(processed_tokens, n_value_tri)
print(f"{n_value_tri}-grams (Trigrams):", trigrams_list)


# --- Step 4: Compute N-gram Frequencies ---

print("\n--- Computing Frequencies ---")
# Unigram frequencies (needed for denominators and context)
unigrams_list = generate_ngrams(processed_tokens, 1)
unigram_freqs = Counter(unigrams_list)
print("Unigram Frequencies:", unigram_freqs)

# Bigram frequencies
bigram_freqs = Counter(ngrams_list) # Use the ngrams_list from Step 3
print("\nBigram Frequencies:", bigram_freqs)
print("\nMost Common Bigrams:", bigram_freqs.most_common(5))

# Trigram frequencies
trigram_freqs = Counter(trigrams_list)
print("\nTrigram Frequencies:", trigram_freqs)
print("\nMost Common Trigrams:", trigram_freqs.most_common(5))


# --- Step 5: Build a Probabilistic Language Model (Bigram with Add-1 Smoothing) ---

print("\n--- Probabilistic Model (Bigram with Add-1 Smoothing) ---")

vocabulary = set(processed_tokens)
vocab_size = len(vocabulary)
print(f"Vocabulary Size (V): {vocab_size}")

def calculate_smoothed_prob(prev_ngram, target_word, n_minus_1_gram_freqs, n_gram_freqs, vocab_size, k=1):
    """
    Calculates Add-k smoothed probability P(target_word | prev_ngram).
    For bigrams: prev_ngram is a unigram tuple, e.g., ('word',) target_word is the next word string
    """
    # Construct the full n-gram tuple we are interested in
    full_ngram = prev_ngram + (target_word,)

    # Get counts
    ngram_count = n_gram_freqs.get(full_ngram, 0)
    prev_ngram_count = n_minus_1_gram_freqs.get(prev_ngram, 0) # Use n-1 gram frequency for denominator

    # Apply Add-k smoothing
    probability = (ngram_count + k) / (prev_ngram_count + k * vocab_size)
    return probability

# Example: Calculate P('modeling' | 'ngram')
prev_ngram_example = ('ngram',)
target_word_example = 'modeling'

prob_example = calculate_smoothed_prob(
    prev_ngram_example,
    target_word_example,
    unigram_freqs, # N-1 gram freqs (unigrams for bigram model)
    bigram_freqs,  # N gram freqs (bigrams for bigram model)
    vocab_size,
    k=1            # Add-1 smoothing (Laplace)
)

print(f"Smoothed P('{target_word_example}' | '{prev_ngram_example[0]}') = {prob_example:.4f}")

# Example: Calculate P('structure' | 'language')
prob_example_2 = calculate_smoothed_prob(
    ('language',), 'structure', unigram_freqs, bigram_freqs, vocab_size, k=1
)
print(f"Smoothed P('structure' | 'language') = {prob_example_2:.4f}")

# Example: Calculate P('analysis' | 'text') - unseen in this tiny corpus?
prob_example_3 = calculate_smoothed_prob(
    ('text',), 'analysis', unigram_freqs, bigram_freqs, vocab_size, k=1
)
print(f"Smoothed P('analysis' | 'text') = {prob_example_3:.4f}") # Should be > 0 due to smoothing


# --- Step 6: Generate a Transition Matrix (Bigram Probabilities) ---

print("\n--- Transition Matrix (Bigram Smoothed Probabilities) ---")

# Using a dictionary of dictionaries for sparse representation
transition_matrix = defaultdict(dict)
vocabulary_list = sorted(list(vocabulary)) # Get a sorted list for consistent matrix structure

for prev_word_token in vocabulary_list:
    prev_unigram = (prev_word_token,) # N-1 gram context
    for current_word_token in vocabulary_list:
        prob = calculate_smoothed_prob(
            prev_unigram,
            current_word_token,
            unigram_freqs,
            bigram_freqs,
            vocab_size,
            k=1
        )
        transition_matrix[prev_word_token][current_word_token] = prob

# Print a portion of the matrix for demonstration
print("Example transitions (rows are 'from', columns are 'to'):")
words_to_show = ['ngram', 'language', 'text', 'modeling']

# Print header
print(' '*15, end='')
for word in words_to_show:
    print(f"{word:<15}", end="")
print()

# Print rows
for from_word in words_to_show:
    if from_word in transition_matrix:
        print(f"{from_word:<15}", end="")
        for to_word in words_to_show:
            prob = transition_matrix[from_word].get(to_word, 0.0) # Get prob or 0 if word somehow missed
            print(f"{prob:<15.4f}", end="")
        print()

# --- End of Example ---
print("\nN-gram analysis complete.")
```

```python
import re
from collections import Counter
from nltk.util import ngrams

# --- Step 1: Define Sample Text ---
text = '''
N-gram modeling is a fundamental technique.
Text analysis benefits from n-gram models.
Understanding n-grams helps in understanding language structure.
Probabilistic models use n-gram frequencies.
'''

# --- Step 2: Preprocess Text ---
def preprocess(text):
    text = re.sub(r'[^\w\s]', '', text.lower())
    return text.split()

tokens = preprocess(text)
vocab = set(tokens)
vocab_size = len(vocab)

# --- Step 3: Generate N-grams ---

unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

# --- Step 4: Frequency Counts ---
unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

for ugfreq in unigram_freq.items():
    print(*ugfreq[0], ':', ugfreq[1])
for bgfreq in bigram_freq.items():
    print(*bgfreq[0], ':', bgfreq[1])
for tgfreq in trigram_freq.items():
    print(*tgfreq[0], ':', tgfreq[1])

print(f'\nTop-3 most comman unigram are : {unigram_freq.most_common(3)}')
print(f'Top-3 most comman bigram are : {bigram_freq.most_common(3)}')
print(f'Top-3 most comman trigram are : {trigram_freq.most_common(3)}')

# --- Step 5: (Laplace) Add-1 Smoothed Probability ---
def smoothed_prob(prev, word, n1_freq, n_freq, V, k=1):
    bigram = prev + (word,)
    return round((n_freq[bigram] + k) / (n1_freq[prev] + k * V), 5)

# --- Step 6: Example Probabilities ---
print('\n--- Smoothed Probabilities ---')
print("P('modeling' | 'ngram') =", smoothed_prob(('ngram',), 'modeling', unigram_freq, bigram_freq, vocab_size))
print("P('structure' | 'language') =", smoothed_prob(('language',), 'structure', unigram_freq, bigram_freq, vocab_size))
print("P('analysis' | 'text') =", smoothed_prob(('text',), 'analysis', unigram_freq, bigram_freq, vocab_size))

# --- Step 7: Transition Matrix (Example) ---
print('\n--- Transition Matrix (partial) ---')
sample_words = ['ngram', 'language', 'text', 'modeling']

print(f'{'':15}', end='')
for w in sample_words:
    print(f'{w:15}', end='')
print()

for w1 in sample_words:
    print(f'{w1:15}', end='')
    for w2 in sample_words:
        prob = smoothed_prob((w1,), w2, unigram_freq, bigram_freq, vocab_size)
        print(f'{prob:<15}', end='')
    print()
```
