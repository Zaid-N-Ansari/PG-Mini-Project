```python
from collections import Counter
from nltk.corpus import stopwords
```

```python
fp = open('nlp_text.txt', 'r')
corpus:str = ''
for line in fp.readlines():
    corpus += line

corpus = corpus.replace('\n', '').replace(',', '').replace('.', '').lower()
corpus_words = corpus.split()
corpus_words_count = Counter(corpus_words)
corpus_words_count.values()
corpus_words_count=[itm for itm in corpus_words_count.items()]
corpus_words_count.sort(key=lambda x: x[1], reverse=True)
stop_words = set(stopwords.words('english'))
stop_word = [t for t in corpus_words_count if t[0] in stop_words]
print(*[f'-- {_[0]} : {_[1]} --' if _ in stop_word else f'{_[0]} : {_[1]}' for _ in corpus_words_count], sep='\n')
```

```python
def LevenshteinDistance(s1:str, s2:str):
    m, n = len(s1), len(s2)
    w_ins, w_del, w_sub = 1, 1, 1
    word_nest = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    s1, s2 = s1.lower(), s2.lower()

    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                word_nest[i][j] = j * w_ins
            elif j == 0:
                word_nest[i][j] = i * w_del
            else:
                cost = 0 if s1[i - 1] == s2[j - 1] else w_sub
                word_nest[i][j] = min(
                    word_nest[i - 1][j] + w_del,
                    word_nest[i][j - 1] + w_ins,
                    word_nest[i - 1][j - 1] + cost
                )

    return word_nest[m][n]

str1 = 'grown'
str2 = 'known'

distance = LevenshteinDistance(str1, str2)
print(f'Levenshtein Distance: {distance}')
```
